```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
folder= Path("data")
pattern = "resultados_analogos_*.xlsx"
df_list=[]
```

# cluster

```python
cluster = pd.read_excel('data/analogos_years_icc_stations.xlsx')
cluster
```

```python
cluster_ = cluster.reset_index().melt(id_vars='index', var_name='year', value_name='year_ana').drop(columns='index')
cluster_['year_ana']= cluster_['year_ana'].astype(int)
```

```python
result=[]
anos=[]
estaciones = []
for station in iell.station.unique():
    #print(station)
    for year in cluster.columns.to_list():
        mean = 0
        for i in [0, 1, 2, 3, 4]:
            star = iell[(iell.station==station) & (iell.year==int(cluster[year][i]))]
            if not star.empty:
                mean=mean + star.sem_iell.values[0]
        result.append(float(mean/5))
        anos.append(year)
        estaciones.append(station)
    
cluster_m = pd.DataFrame({
    'station': estaciones,
    'year': anos,
    'cluster': result
})
cluster_m 

```

```python
new_m = all_info[all_info['_merge']=='both'][['year','month','escenario','station','semana_promedio','semana_ponderada','sem_iell_real']].reset_index(drop=True)
new_m.rename(columns={'semana_promedio':'prom', 'semana_ponderada':'pd'}, inplace=True)
```

```python
df_pivot = new_m.pivot_table(
    index=['year','station'],
    columns='escenario',
    values=['prom', 'pd'] # Especificamos ambas columnas que queremos pivotar
)
df_pivot.columns = [f"{col_name[1]}_{col_name[0]}" for col_name in df_pivot.columns]
df_pivot = df_pivot.reset_index()
complete_info = df_pivot.merge(new_m[['year', 'station', 'sem_iell_real']].drop_duplicates().reset_index(drop=True), how='inner', on=['year','station']).merge(cluster_m, how='inner', on=['year','station'])
complete_info
```

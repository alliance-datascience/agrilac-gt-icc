```python
input_path= "D:\\OneDrive - CGIAR\\agrilac\\icc\\data\\PUERTOSANJOSE.xlsx"

station = pd.read_excel(input_path)
station.head()
```

```python
monthly_prec =station.groupby(['year', 'month'], as_index=False)['prec'].sum()
```

```python
#===========input generation
name_station = os.path.basename(input_path).lower()
output_name = name_station.replace('.xlsx', '_prec.xlsx')
output_path=os.path.join(os.path.dirname(input_path), output_name)
monthly_prec.to_excel(output_path, index=False)
```
